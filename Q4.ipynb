```python
import pandas as pd
import re

# 读取作者信息数据
file_path = r"C:\Users\z\Desktop\数据挖掘\数据挖掘指定任务\论文作者信息.xlsx"
df = pd.read_excel(file_path)

# 定义清洗函数
def clean_author_name(author_name):
    # 去掉末尾的点
    author_name = re.sub(r'\.$', '', author_name)
    # 去掉末尾的数字
    author_name = re.sub(r'\d+$', '', author_name)
    return author_name

# 清洗每一列的作者名称
for col in df.columns:
    df[col] = df[col].apply(lambda x: clean_author_name(str(x)) if pd.notna(x) else x)

# 查看清洗后的结果
print(df)

# 如果需要，可以保存到新的文件
output_path = r"C:\Users\z\Desktop\清洗后的论文作者信息.xlsx"
df.to_excel(output_path, index=False)
```
